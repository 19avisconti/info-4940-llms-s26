# Compare Base and Instruction-Tuned Models

In the pre-training phase, a model is trained on large collections of naturally occurring text from the web. After this phase, the model has learned general language patterns and can generate plausible continuations of a given prompt, similar to how text appears online. This is referred to as a **base model**.

### Limits of the Base Model

However, we often want models to interact in ways that go beyond the patterns found in generic web data.

While a base model has learned general language patterns from massive amounts of raw text, it **hasn’t yet been adjusted** to follow instructions or hold conversations.

It might respond to a question **with another question** or generate text that reads more like a **Wikipedia article** than a conversational exchange.

### Fine-Tuning Phase

In the **fine-tuning phase**, the base model is trained on **specialized datasets** that contain training examples. These examples are tokenized sequences consisting of **inputs** and their **corresponding desired outputs**.

Each training example is an **input–output pair**:

* The **input** is a sample of text provided to the model (such as a sentence).
* The **output** is the expected response it should learn to generate.

**Note:** The sample output is often called a **label**, and therefore the training example is often referred to as a **labeled example**.

### Purpose of Fine-Tuning

This fine-tuning phase trains the model to produce outputs in **specific styles or structures** — for example:

* Answering questions directly
* Summarizing text
* Writing code
* Engaging in natural conversation

A chatbot, for instance, is fine-tuned on dialogue-style data that shows how to respond helpfully and politely to user prompts. Fine-tuned models may also include **locally relevant** or **domain-specific** information, such as medical terminology or customer service language

Fine-tuning uses **far less data** than pre-training and is **tailored to a particular purpose**.

### Supervised vs. Unsupervised

It is worth noting that the **pre-training phase** is often called **unsupervised pre-training**, because the model learns directly from **raw text** rather than from structured input–output pairs.

The **fine-tuning phase**, by contrast, is often called **supervised training**, since it relies on **labeled examples** — input–output pairs that specify the desired behavior.

## Load Base and Instruction-Tuned Models

Run the code cell below to import the necessary packages and load both the base model and the instruction-tuned model, along with their corresponding tokenizers.

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

base_model_name = "Qwen/Qwen2.5-0.5B"
base_model = AutoModelForCausalLM.from_pretrained(base_model_name)
base_tokenizer = AutoTokenizer.from_pretrained(base_model_name)

instruct_model_name = "Qwen/Qwen2.5-0.5B-Instruct"
instruct_model = AutoModelForCausalLM.from_pretrained(instruct_model_name)
instruct_tokenizer = AutoTokenizer.from_pretrained(instruct_model_name)
```

## Define a Text Generation Function

This code cell below defines a simple function for generating text. The function `generate_text()` takes a text prompt, tokenizes it, and uses the model to generate a continuation consisting of up to 100 tokens. It returns the original prompt together with the generated continuation.

```python
def generate_text(prompt, model, tokenizer):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(**inputs, max_new_tokens=100,  pad_token_id = tokenizer.eos_token_id)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

Next, use the `generate_text()` function to generate continuations for your prompt using the base model and its associated tokenizer. Print the results.

```python
# Generate text with base model
result_base = generate_text(prompt, base_model, base_tokenizer)
print(result_base)
```

Now that you have generated text from the base model, use the `generate_text()` function to generate continuations for your prompt using the instruction-tuned model and its associated tokenizer. Print the results.

```python
# Generate text with instruction-tuned model
result_instruct = generate_text(prompt, instruct_model, instruct_tokenizer)
print(result_instruct)
```
