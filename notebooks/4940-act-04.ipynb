```python
from datasets import load_dataset
import pandas as pd

# Load a very small subsample of the larger dataset. 
pretraining_subset = load_dataset("HuggingFaceFW/fineweb", "sample-10BT", split="train", streaming=True).take(10000)

pretraining_df = pd.DataFrame(pretraining_subset)
```

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
import seaborn as sns


cv = CountVectorizer(stop_words="english", max_features=1024)
vecs = cv.fit_transform(pretraining_df.text)
id2token = {v: k for k, v in cv.vocabulary_.items()}
```

```python
k = 20

total_words = vecs.sum()
total_per_word = np.array(vecs.sum(axis=0))[0]
top_k_idx = np.argsort(-total_per_word)[:k]
top_k_freqs = (total_per_word / total_words)[top_k_idx]
top_k_words = [id2token[idx] for idx in top_k_idx]

for word, freq in zip(top_k_words, top_k_freqs):
    print(f"{word}: {freq:.3f}")
```
